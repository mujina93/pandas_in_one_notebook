# Custom sorting in a series

```python
# create a simple DataFrame, with 2 columns. We would like to be able to order according to the second column. But since
# those are strings, python doesn't know how to order according to them, other then going for alphabetical order!
df = pd.DataFrame({'ID':[100,101,102,103], 'quality':['good','very good','good','excellent']})
```

```python
df
```

```python
# sorting sorts by alphabetical order
df.sort_values('quality')
```

```python
# if we store the series as a category, we can assign an explicit ordering of the category
df['quality'] = df.quality.astype('category',categories=['good','very good','excellent'], ordered=True)
```

```python
# from pyhton3.6, write the cast to 'category' like this:
df['quality'] = df.quality.astype(pd.api.types.CategoricalDtype(categories=['good','very good','excellent'], ordered=True))
```

```python
# now python knows that is has to use the order: good < very good < excellent !
df.sort_values('quality')
```

```python
# we can also filter using comparison operators on our custom (strings) values!
df.loc[df.quality > 'good', :]
```

# Dummy (or Indicator) variables

```python
# titanic dataframe from kaggle
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
# dummy-encoding of a column. Takes the column Sex (which has values 'female' or 'male', 
# and create a new column (Sex_male), in which 'female' is substituted with 0 and 'male with 1.
# this column stores now the indicator variable of the event "this person is a male"
train['Sex_male'] = train.Sex.map({'female':0, 'male':1})
```

# SettingWithCopyWarning

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
# counting nulls
movies.content_rating.isnull().sum()
```

```python
# rows with missing content_rating
movies[movies.content_rating.isnull()]
```

```python
# representing a value with pandas' NaN
# here, we want to set the movies with 'NOT RATED' rating to NaN
movies.content_rating.value_counts()
```

```python
import numpy as np
```

```python
# NaN is numpy.nan. We change all the content_rating entries of the rows which have content_rating=='NOT RATED' to NaN
movies[movies.content_rating == 'NOT RATED'].content_rating = np.nan
```

```python
movies.content_rating.isnull().sum()
```

```python
# warning!
```

```python
# solution: using loc
# remember: use loc, don't filter rows and then select columns, as we did above
# the problem is that the operation of accessing rows might return a copy of the dataset, therefore, once you later
# access the columns, you are accessing the columns of a copy! Instead, loc does all the subsetting in one operation
movies.loc[movies.content_rating == 'NOT RATED', 'content_rating'] = np.nan
```

```python
movies.content_rating.isnull().sum()
```

```python
top_movies = movies.loc[movies.star_rating >= 9, :]
```

```python
top_movies
```

```python
top_movies.loc[0, 'duration'] = 150
```

```python
# warning! Why? We are using loc! The problem is that pandas does not know if top_movies is a view or a copy of your
# DataFrame!
# Solution: when you take a view (subsetting, such as above when we defined top_movies), add .copy() at the end, so
# that you are explicitly telling pandas to create a copy (a concrete object, not just a view) of the subset that
# you are filtering out. If you then work on the copy, you won't get any warnings, because pandas will know that it
# has to work on a concrete object, and has no doubt about whether it should modify the original dataset or not.
top_movies_concrete = movies.loc[movies.star_rating >= 9, :].copy()
top_movies_concrete.loc[0, 'duration'] = 150
```
